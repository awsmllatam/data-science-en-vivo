## Top artist with hits songs

```python
artist = df['artist'].value_counts()
artist.head(10)
```

```python
tp_artist_songs = artist[:5]
tp_artist_name = artist[:5].index
fig = plt.figure(figsize = (10,5))
plt.bar(tp_artist_name, tp_artist_songs, width = 0.4, color="forestgreen")
plt.xlabel("Artist")
plt.ylabel("Nro of songs")
plt.title("Top Artist with hit songs")
plt.show()
```

## Averange duration of top hit songs for each year

```python
# Convertir de milisegundos a minutos y segundos
def time_convert(ms):
    sec = ms/1000
    return f"{int(sec//60)}:{int(sec%60)}"

duration = df[['duration_ms','year']].groupby('year').mean().reset_index()
duration['min:sec'] = duration['duration_ms'].apply(time_convert)
duration
```

```python
x = df['year'].unique()
x = sorted(x)
x=pd.Series(x)
y = duration['min:sec']
plt.figure(figsize=(12,8))
sns.lineplot(x,y,data=df)
plt.title('Change in duration songs')
plt.xlabel('Years', size=10)
plt.ylabel('Time', size=10)
plt.show()
```

## Amount of explicit content in top-hits

```python
data = df['explicit'].value_counts()
plt.figure(figsize=(10,8))
plt.title('Total amount of explicit sounds')
plt.ylabel('Nro of songs')
plt.xlabel('Explicitness')
data.plot(kind='bar',color=['darkgreen','greenyellow'])
plt.show()
```

## The most popular song to ever become top-hit

```python
maxpop = df.loc[df['popularity'] == df['popularity'].max()]
maxpop
```

## The most popular artist

```python
pop_artist = df.groupby('artist')[['artist','popularity']].sum().sort_values('popularity',ascending=False).head(10)
pop_artist
```

```python
plt.figure(figsize=(14,8))
plt.title('Most popular artist', fontsize= 15)
c1 = sns.barplot(x='popularity', y=pop_artist.index, data=pop_artist, palette="Greens_r")
c1.bar_label(c1.containers[0], size = 15)
plt.show()
```
